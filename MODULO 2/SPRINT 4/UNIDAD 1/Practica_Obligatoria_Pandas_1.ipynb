La recaudación de Monsters University fue de 754 millones.
Hay 2 nombres que han sufrido el efecto del conocido "error Jaime", corrígelo en todas las series (recuerda que los indices son inmutables, tendrás que hacer algo de código): 


```python
import numpy as np
import pandas as pd
titulos = ["Toy Story 4", "Los Incribles 2", "Buscando a Dory", "Toy Story 3", "Caco", "Inside Out", "Monsters University", "Up"]
lanzamiento = [2015, 2018, 2016, 2010, 2017, 2015, 2013, 2006]
recaudaciones = [1073, 1242, 1029, 1067, 807, 857, 744, 735]  # En millones de dólares
espectadores = [74.91, 93.42, 76.72, 81.35, 62.75, 68.27, 54.74, 54.34] # En millones, estimación hecha por aficionados

#1. creamos las tres series
s1 = pd.Series(lanzamiento, index=titulos)
s2 = pd.Series(recaudaciones, index=titulos)
s3 = pd.Series(espectadores, index=titulos)
print("Las tres series creadas en funcion del lanzamiento, recaudaciones y espectadores", s1, s2, s3)

#2. ordenamos de forma ascendente una serie de recaudaciones
s2_sorted = s2.sort_values(ascending=True)
print("Lista de peliculas ordenada de forma ascendente en funcion de sus recaudaciones", s2_sorted)

#3. usamos un bucle for para mostrar la recaudacion, año de lanzamiento y numero de espectadores de cada pelicula
for titulo, recaudacion in s2_sorted.items():
    lanzamiento = s1[titulo]
    espectadores = s3[titulo]
    precio_medio = recaudacion / espectadores #mostramos el precio medio de la entrada de cada pelicula
    print(titulo,": año de lanzamiento:", lanzamiento, ", Recaudacion:", recaudacion, "millones de dólares", ",", espectadores, "millones de espectadores", ", precio medio de entrada:", precio_medio, "dólares")

#4. contestamos a varias preguntas
pelicula_menor_recaudacion = s2.idxmin() #averiguamos cual es la pelicula con menor recaudacion
print("La pelicula con menor recaudacion es", pelicula_menor_recaudacion)
pelicula_mayor_recaudacion = s2.idxmax() #averiguamos la pelicula con mayor recaudacion
print("La pelicula con mayor recaudacion es", pelicula_mayor_recaudacion)

#averiguamos la pelicula con mas de 1000 millones de recaudacion y menos de 75 millones de espectadores
pelicula = s2[(s2 > 1000) & (s3 < 75)].idxmax() 
print("La pelicula con mas de 1000 millones de recaudacion y menos de 75 millones de espectadores es", pelicula)

#calculamos la edad de Gonzalo en el estreno de cada pelicula si ahora tiene 30 años
edad_gonzalo_estreno = 30 - (2023 - s1) #averiguamos la edad restandole a la edad actual los años que han pasado desde que se estreno la pelicula hasta ahora
for titulo, edad in edad_gonzalo_estreno.items():
    print("Gonzalo tenía", edad, "años cuando se estrenó", titulo)
    
#corregimos errores

s2["Monsters University"] = 754 #corregimos la recaudacion de Monsters University
 
s1["Toy Story 4"] = 2019 #corregimos el año de estreno de Toy Story 4

#corregimos los "errores Jaime"
s1.rename(index={"Caco": "Coco", "Los Incribles 2": "Los Increibles 2"}, inplace=True)
s2.rename(index={"Caco": "Coco", "Los Incribles 2": "Los Increibles 2"}, inplace=True)
s3.rename(index={"Caco": "Coco", "Los Incribles 2": "Los Increibles 2"}, inplace=True)
print(s1, s2, s3)
```
