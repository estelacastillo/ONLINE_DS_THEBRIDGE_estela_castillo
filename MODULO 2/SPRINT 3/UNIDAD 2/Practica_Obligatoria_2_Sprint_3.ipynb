,E,O).

Por ejemplo, si tu posicion inicial fuese (5,0) y la orientación O, no podrias ponerlo porque se sale del tablero. Ahora bien, si fuese orientación este, si podrías, y ocuparía las posiciones [(5,0), (5,1), (5,2), (5,3)]. Por tanto tendrás que iterar con un bucle si es posible encajar el barco en el tablero, y el bucle terminará cuando lo consigas encajar. Ojo con las posiciones donde ya hay barco!!!

Usa import random para elegir las coordenadas aleatorias.

NOTA: Intenta crear fuciones para crear el tablero (de forma que tenga dimensiones variables), posicinar los barcos, generar los disparos dadas unas coordenadas, y generar barcos de forma aleatoria

```python
import numpy as np
tablero = np.full((10, 10), " ") #creamos un array de 10x10 con espacios
tablero[0, 1] = "Y" #posicionamos el primer barco en (0, 1)
tablero[1, 1] = "Y" #posicionamos el primer barco en (1, 1)
tablero[1, 3:7] = "Y" #posicionamos el segundo barco en (1, 3), (1, 4), (1, 5), (1, 6)
print(tablero)
tablero[1, 1] = "X" #recibe un disparo en el primer barco en (1, 1)
print(tablero)
tablero[3, 2] = "A" #recibe un disparo en agua en (2, 2)
print(tablero)
#Ignorando los disparos recibidos, ahora prueba a posicionar otro barco de 4 posiciones dentro del tablero, pero de manera ALEATORIA.
import random
barco_valido = False #variable que controla si el barco es válido
while not barco_valido: 
    fila = random.choice(range(10)) #elegimos una fila al azar entre 0 y 9
    columna = random.choice(range(10)) #elegimos una columna al azar entre 0 y 9
    orientacion = random.choice(["Norte", "Sur", "Este", "Oeste"]) #elegimos una orientación al azar entre norte, sur, este u oeste
    if orientacion == "Norte": 
        if fila >= 3: #comprobamos si el barco cabe en el tablero
            if np.all(tablero[fila-3:fila+1, columna] == " "): #comprobamos si el barco no se solapa con otro
                tablero[fila-3:fila+1, columna] = "Y" #posicionamos el barco
                barco_valido = True #indica que el barco es válido
    elif orientacion == "Sur": 
        if fila <= 6: #comprobamos si el barco cabe en el tablero
            if np.all(tablero[fila:fila+4, columna] == " "): #comprobamos si el barco no se solapa con otro
                tablero[fila:fila+4, columna] = "Y" #posicionamos el barco
                barco_valido = True #indica que el barco es válido
    elif orientacion == "Este": 
        if columna <= 6: #comprobamos si el barco cabe en el tablero
            if np.all(tablero[fila, columna:columna+4] == " "): #com`robamos si el barco no se solapa con otro
                tablero[fila, columna:columna+4] = "Y" #posicionamos el barco
                barco_valido = True #indica que el barco es válido
    elif orientacion == "Oeste": 
        if columna >= 3: #comprobamos si el barco cabe en el tablero
            if np.all(tablero[fila, columna-3:columna+1] == " "): #comprobamos si el barco no se solapa con otro
                tablero[fila, columna-3:columna+1] = "Y" #posicionamos el barco
                barco_valido = True #indica que el barco es válido
print(tablero)
```
